```python
"""

Created on Friday October 18 11:47:42 2018
@author: Kanat.Osmonov
"""

"""
DocString:

A) Introduction:
This game is based on house escape quest. It consists of four stages, which are four rooms.
The game has defined functions for each room, start, and fail. It's an easy game and requires
common logic, however, there is very tricky question in the end that requires specific 
mathematical knowledge.

room_1: Take an action to proceed further
room_2: Decide which object to interact with
room_3: Get out of the dark room
room_4: Give mathematically justified response

B) Known Bugs and/or Errors:
None.

"""


from sys import exit
import time
import IPython
from random import randint


def game_start():
    global player
    print("\nHello, we welcome you in our HOUSE ESCAPE quest today!\nWe are super excited to host you here!\n")
    player = input("Please introduce yourself: \n")
    print(f"Thank you {player}. So now, you will begin your escape quest from our house. Best of luck!!!")
    room_1()
    

def room_1():
    trials = 0 
    rand_value = str(randint(1,5))

    choice = input("""
You see a door with a locker on it.\n 
What is your action?
You can choose one from the following:
1) Kick the door to open it
2) Guess a locker code of 1 digit
3) Look for other routes  \n""").lower()

    if "1" in choice or "kick" in choice:
        print("You’ve got a weak kick, the door won’t open")
        room_1()
    elif "3" in choice or "look" in choice or "other" in choice or "routes" in choice:
        print("Looks like there is only one route – DOOR!")
        room_1()
    elif "2" in choice or "guess" in choice or "locker" in choice:
            while trials < 4:
                guess = input("\nWrite your guess between 1 and 5 inclusive in numberic value: \n")
                if guess == rand_value:
                    print("You are absolutely right, best of luck in the next room!")
                    input('Press any key to enter the next room\n')
                    room_2()
                    break
                elif guess != rand_value:
                    print("\nSorry, that's incorrect. Make another guess")
                    trials += 1
                    if trials == 3 :
                        fail()
                else: 
                    print("You typed something incorrectly, try again")
                    room_1()
                 
    else:
        print("You typed something incorrectly, try again")
        room_1()
        
        
def room_2():       
    print("""You've entered into the second room. Here you can observe a chair in the left corner, a bag \nin the middle of the room, and a picture hanging on the right wall.""")
    action = input("What action would you like to take? \nWrite your answer below:\n").lower()
    if 'bag' in action:
        print("Looks like someone emptied it before you. Do something else!")
        time.sleep(3)
        room_2()
    elif 'chair' in action:
        print("You can only have a rest on the chair, then look somewhere else!")
        time.sleep(3)
        room_2()
    elif 'move' and 'picture' in action or 'take' and 'picture' in action:
        button = input("""\nYou’ve found a big button behind the picture. What would you like to do with it?
\nWrite your response below:\n""").lower()
        if "push" in button or "press" in button or "punch" in button:
            print("The secret door has opened to another room and you are entering in there!")
            time.sleep(2)
            room_3()
        else:
            fail()
    else: 
        print("You typed something incorrectly, please try again")
        room_2()

        
def room_3():
    action_2 = input("""You have entered into a very dark room. Nothing to see.
    You have the following options:
    1) Look for a switch
    2) Turn a flashlight in your phone
    3) Give up on a game and call for a help because you freak out from the darkness
    What action do you take?\n""").lower()
    if "2" in action_2 or "flashlight" in action_2 or "phone" in action_2:
        print("Oh my God, looks like your battery is dead. Always keep it charged")
        time.sleep(2)
        room_3()
    elif "3" in action_2 or "give up" in action_2 or "help" in action_2 or "quit" in action_2:
        fail()
    elif "1" in action_2 or"switch" in action_2 or "look" in action_2:
        time.sleep(4)
        print("Oh my God, it's very hard to look for something in the dark")
        time.sleep(4)
        print("Can't find the switch. Maybe giving up was a good option???")
        time.sleep(4)
        print("Finally you find it and turn it on. Now, you can see a a door in the far left corner and enter it")
        time.sleep(2)
        input("Press any key to enter the next room")
        room_4()
    else:
        print("You've typed something wrong, please try again")
        room_3()
            

def room_4():        
    action_3 = input("""You've made it so far. And this is the final room. Best of LUCK!!!\nSurprisingly, you see three doors here: green, yellow, and orange. \nYou should choose a correct door.\nWhich door do you choose to enter?\n
    """).lower()
    if 'green' in action_3:
        green = input("""Actually, we decided to make the job easier for you and we are opening the yellow door for you. 
        And, as you see, there was a wall behind it. Now, you have only two options: would you like to keep 
        your GREEN door option or switch to the ORANGE one?
        NOTE: Make your answer best in mathematical terms. If you don't know what it means, go with your intuition.\n""").lower()
        if 'keep' in green or 'green' in green:
            print("This was an incorrect choice. You will definitely succeed after reading Monty Hall problem")
            fail()
        elif 'switch' in green or 'orange' in green:
            print("You are very smart. Congratulations!!! YOU WON THE GAME")
        else:
            print("You typed something wrong, try again")
            room_4()
            
    elif 'yellow' in action_3:
        yellow = input("""Actually, we decided to make the job easier for you and we are opening the GREEN door for you. 
        And, as you see, there was a wall behind it. Now, you have only two options: would you like to keep 
        your yellow door option or switch to the ORANGE one?
        NOTE: Make your answer best in mathematical terms. If you don't know what it means, go with your intuitio
        n.\n""").lower()
        if 'keep' in yellow or 'yellow' in yellow:
            print("This was an incorrect choice. You will definitely succeed after reading Monty Hall problem")
        elif 'switch' in yellow or 'orange' in yellow:
            print("You are very smart. Congratulations!!! YOU WON THE GAME")
        else:
            print("You typed something wrong, try again")
            room_4()

    elif 'orange' in action_3:
        orange = input("""Actually, we decided to make the job easier for you and we are opening the YELLOW door for you. 
        And, as you see, there was a wall behind it. Now, you have only two options: would you like to keep 
        your ORANGE door option or switch to the GREEN one?
        NOTE: Make your answer best in mathematical terms. If you don't know what it means, go with
        your intuition.\n""").lower()
        if 'keep' in orange or 'orange' in orange:
            print("This was an incorrect choice. You will definitely succeed after reading Monty Hall problem")
        elif 'switch' in orange or 'green' in orange:
            print("You are very smart. Congratulations!!! YOU WON THE GAME")
        else:
            print("You typed something wrong, try again")
            room_4()
    else:
        print("You typed something wrong, try again")
        room_4()
        

def fail():
    print(f"We are sad that you didn't make it through {player}, you will definitely succeed next time!")
    
    response = input(f"{player} would you like to play one more time? (Yes/No)").lower()
    if response == 'yes':
        room_1()
        
    else:
        print(f"Thanks for playing {player}. See you next timme!")
        exit(0)

game_start()


```
